```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import svm
import time

data = pd.read_csv("lego.csv", header=None)

# Datos CSV
X = data[[0,1]].values
d = data[[2]].values

# Datos random para probar que tanto ajusta el modelo
ex = [[np.random.random(), np.random.random()] for _ in range(1000)]
```

```python
def see_result(X,d):
    a1,a2,b1,b2 = [], [], [], []

    for x, di in zip(X,d):
        if di == 1:
            a1.append(x[0])
            b1.append(x[1])
        else:
            a2.append(x[0])
            b2.append(x[1])
    return (a1,b1,a2,b2)

x1,y1,x2,y2 = see_result(X,d)

plt.plot(x1,y1, 'ro', x2,y2, '^b')
plt.title("Datos lego")
plt.show()
```

### Pruebas con kernel gaussiano

```python
start = time.time()
clt1 = svm.SVC(kernel='rbf', C=1000, gamma='scale')
clt1.fit(X, d.ravel())
print(f"Tiempo: {time.time() - start}seg\n")
print("Ajuste de los datos con el modelo", clt1.score(X,d.ravel()))

result1 = clt1.predict(ex)
x11,y11,x21,y21 = see_result(ex,result1)
```

```python
start = time.time()
clt2 = svm.SVC(kernel='rbf', C=100000, gamma='scale')
clt2.fit(X, d.ravel())
print(f"Tiempo: {time.time() - start}seg\n")
print("Ajuste de los datos con el modelo", clt2.score(X,d.ravel()))

result2 = clt2.predict(ex)
x12,y12,x22,y22 = see_result(ex,result2)
```

```python
start = time.time()
clt3 = svm.SVC(kernel='rbf', C=1, gamma='scale')
clt3.fit(X, d.ravel())
print(f"Tiempo: {time.time() - start}seg\n")
print("Ajuste de los datos con el modelo", clt3.score(X,d.ravel()))

result3 = clt3.predict(ex)
x13,y13,x23,y23 = see_result(ex,result3)
```

```python
start = time.time()
clt5 = svm.SVC(kernel='rbf', C=0.01, gamma='scale')
clt5.fit(X, d.ravel())
print(f"Tiempo: {time.time() - start}seg\n")
print("Ajuste de los datos con el modelo", clt5.score(X,d.ravel()))

result5 = clt5.predict(ex)
x15,y15,x25,y25 = see_result(ex,result5)
```

```python
plt.subplot(231)
plt.plot(x11,y11, 'ro', x21,y21, '^b')
plt.title("C: 1000")

plt.subplot(232)
plt.plot(x12,y12, 'ro', x22,y22, '^b')
plt.title("C: 100000")

plt.subplot(233)
plt.plot(x13,y13, 'ro', x23,y23, '^b')
plt.title("C: 1")

plt.subplot(234)
plt.plot(x14,y14, 'ro', x24,y24, '^b')
plt.title("C: 0.1")

plt.subplot(235)
plt.plot(x15,y15, 'ro', x25,y25, '^b')
plt.title("C: 0.01")

plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.5,
                    wspace=0.5)

plt.show()
```

### Conclusiones

Entrenamos 2 SVM en general, variando los parametros de cada SVM.

La primera prueba fue con kernel Gaussiano, probamos varios valores para el parametro de regularizacion C.
Tomando como inicial el valor de C a 1000 tuvimos muy buenos resultados, dandonos un ajuste del 99% de los datos, inclusive al probar los datos de random, vemos que la grafica es bastante similar a la grafica de los datos de entrenamiento, al igual que con C a 100000, en esta inclusive se ajusta un poco mas. Ahora cuando empezamos a bajar el valor de C, vemos que la el modelo cae en cuanto a porcentaje de datos bien clasificados, podemos ver que mientras mas pequeo es peor el ajuste de los datos.

La segunda prueba con kernel Polinomial, probamos varios valores para el parametro de regularizacion C ademas de el grado del polinomio en 2 y 6.
Nuevamente, tomamos como inicial C a 1000 pero no tuvimos buenos resultados, incluso aumentando el parametro de regularizacion obtuvimos una mejora pero no fue suficiente como para concluir que se obtuvo una clasificacion bien.
Cuando aumentamos el grado a 6 vemos unas mejoras pero, igualmente no podemos concluir que obtuvimos una buena clasificacion.

Ahora, un punto importante a tratar el tiempo de entrenamiento, para el caso una SVM con kernel Gaussiano el tiempo de entrenamiento fue muy poco, aun asi dando mucho mejores resultados que con un kernel polinomial.
Otro punto, en el caso de la SVM con kernel polinomial vemos que aumentado los grados del polinomio no obtuvimos mejores resultados notables, cumple con la teoria asi que probar con un grado mayor seria innesesario. 
Por ultimo, podemos ver la accion del parametro de regularizacion en el caso del SVM con kernel Gaussiano, aumentando el parametro C podemos ver que efectivamente hay una mayor suavidad en la region de decision, si lo disminuimos vemos que solo abarca donde hay mayor concentracion de puntos, sin embargo, esta entonacion de parametros funciona con este dataset debido a su naturaleza, con lo que este estudio no es regla para generalizar el entrenamiento de una SVM.
